```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import datasets
```

```python
diabetes = datasets.load_diabetes()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.2, random_state=1)
```

```python
x_train.shape, x_test.shape, y_train.shape,y_test.shape
```

```python
alg_diabetes = LinearRegression()
alg_diabetes.fit(x_train, y_train)
```

```python
alg_diabetes.coef_, alg_diabetes.intercept_
```

```python
y_pred = alg_diabetes.predict(x_test)
```

```python
print(alg_diabetes.score(x_test, y_test))
print(alg_diabetes.score(x_train, y_train))
```

```python
print(type(x_test),y_pred)
```

```python
pd.DataFrame(y_pred).to_csv("prediction.csv")
```
